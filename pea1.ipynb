```python
#importing faker and pandas
from faker import Faker
import pandas as pd
import datetime as dt
import random

#Initialize faker
fake = Faker()

#Generating fake data for 500 students
student_data = []

for i in range (0, 500):
    student_data.append({"student_id": i, "name": fake.name(),"age": fake.random_int(min=18, max=50), "gender": fake.random_element(elements=("Male", "Female")), "email": fake.email(), "phone": fake.random_number(digits = 10), "enrollment_year": fake.random_int(min=2010, max=2025), "course_batch": fake.random_element(elements = ("AA-23-23", "BB-55-50", "CC-78-98", "DD-66-67", "EE-90-93")), "city": fake.city(), "graduation_year": fake.random_int(min=1998, max=2025)})
df = pd.DataFrame(student_data)
print(df)
```

```python
#Generating fake data for 500 students on their programming details
programming_data = []

for i in range (0, 500):
    programming_data.append({"programming_id": "PG" + str(i) + "A", "student_id": i, "language": ', '.join(fake.random_elements(elements=("JAVA", "SQL", "Python"), length=random.randint(1,3), unique=True)), "problems_solved": fake.random_int(min=50, max=500), "assessments_completed": fake.random_int(min=5, max=30), "mini_projects": fake.random_int(min=2, max=8), "certification_earned": fake.random_int(min=0, max=30), "latest_project_score": fake.random_int(min=35, max=100)})
df = pd.DataFrame(programming_data)
print(df)
```

```python
#Generating fake data for 500 students on their soft softskills_data
softskills_data = []

for i in range (0, 500):
    softskills_data.append({"soft_skill_id": "SS"+str(i)+"B","student_id": i, "communication": fake.random_int(min=10, max=100), "teamwork": fake.random_int(min=10, max =100), "presentation": fake.random_int(min=10, max=100), "leadership": fake.random_int(min=10, max=100), "critical_thinking":fake.random_int(min=10, max=100), "interpersonal_skills": fake.random_int(min=10, max=100)})
df = pd.DataFrame(softskills_data)
print(df)
```

```python
#Genrating fake data for 500 students on their placement details
placement_data = []

for i in range (0, 500):
    placement_data.append({"placement_id": "PC"+str(i)+"C", "student_id": i, "mock_interview_score": fake.random_int(min=10, max=100), "internships_completed": fake.random_int(min=0, max=10), "placement_status": fake.random_element(elements = ("Ready", "Not Ready", "Placed")), "company_name": fake.random_element(elements = ("ABCD", "EFGH", "IJKL", "MNOP", "QRSTU", "VWXYZ")), "placement_package": fake.random_int(min=40000, max=200000), "interview_rounds_cleared": fake.random_int(min=0, max=4), "placement_date": fake.date_between_dates(date_start=dt.date(2015, 1, 1), date_end=dt.date(2025, 6, 20))})
df = pd.DataFrame(placement_data)
print(df)
```

```python
#Create database if it doesn't exist
cursor_mysql.execute("CREATE DATABASE IF NOT EXISTS Placement;")
```

```python
cursor_mysql.execute("USE Placement;")  # Select database
cursor_mysql.execute("""
    CREATE TABLE IF NOT EXISTS student_data (
        student_id INT PRIMARY KEY,
        name VARCHAR(50),
        age INT,
        gender VARCHAR(50),
        email VARCHAR(50),
        phone VARCHAR(10),
        enrollment_year INT,
        course_batch VARCHAR(10),
        city VARCHAR(50),
        graduation_year INT 
    );
""")
conn_mysql.commit()
print("Table 'student_data' created successfully in MySQL!")
```

```python
cursor_mysql.execute("""
    CREATE TABLE IF NOT EXISTS programming_data (
        programming_id VARCHAR(10) PRIMARY KEY,
        student_id INT,
        language VARCHAR(100),
        problems_solved INT,
        assessments_completed INT,
        mini_projects INT,
        certification_earned INT,
        latest_project_score INT,
        FOREIGN KEY (student_id) REFERENCES student_data(student_id)
    );
""")
conn_mysql.commit()
print("Table 'programming_data' created successfully in MySQL!")
```

```python
cursor_mysql.execute("""
    CREATE TABLE IF NOT EXISTS softskills_data (
        soft_skill_id VARCHAR(10) PRIMARY KEY,
        student_id INT,
        communication INT,
        teamwork INT,
        presentation INT,
        leadership INT,
        critical_thinking INT,
        interpersonal_skills INT,
        FOREIGN KEY (student_id) REFERENCES student_data(student_id)
    );
""")
conn_mysql.commit()
print("Table 'softskills_data' created successfully in MySQL!")
```

```python
cursor_mysql.execute("""
    CREATE TABLE IF NOT EXISTS placement_data (
        placement_id VARCHAR(10) PRIMARY KEY,
        student_id INT,
        mock_interview_score INT,
        internships_completed INT,
        placement_status VARCHAR(50),
        company_name VARCHAR(5),
        placement_package INT,
        interview_rounds_cleared INT,
        placement_date DATE,
        FOREIGN KEY (student_id) REFERENCES student_data(student_id)
    );
""")
conn_mysql.commit()
print("Table 'placment_data' created successfully in MySQL!")
```
